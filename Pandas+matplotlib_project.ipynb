```python
# Accept the class1 percentage class2,class3 as per user

import pandas as pd 

no=int(input("how many student percentage you want")) 
print(f" we want {no} stuent percentage anylysis")
n=no

name=[]
class1_per=[]
class3_per=[]
class2_per=[]

for i in range(n):
    
    nm=input(" Enter the student name here:")
    cls1p=float(input("Enter the class 1 percentage here:"))
    cls2p=float(input("Enter the class 2 percentage here:"))
    cls3p=float(input("Enter the class 3 percentage here:")) 
    
    name.append(nm)
    class1_per.append(cls1p)
    class2_per.append(cls2p)
    class3_per.append(cls3p)
    
data={"Name_of_student":name,"Class1 Percentage":class1_per,"Class2 Percentage":class2_per,"Class3 Percentage":class3_per}

df=pd.DataFrame(data)
df
```

```python
#* Get values from the columns of the DataFrame and plot a Comparison Plot(plot every data in
#  single plot) With different Colors

import matplotlib.pyplot as plt


df.plot(x="Name_of_student", y=["Class1 Percentage", "Class2 Percentage","Class3 Percentage"],kind="bar" ,figsize=(10,8))
plt.legend(loc="upper right")

plt.ylim(10,100)

plt.xlabel("Student Name ",color = "black", weight = "bold", size = 12)
plt.ylabel("Percentage",color = "black", weight = "bold", size = 12)
plt.title("Percentage of student",color = "black", weight = "bold", size = 16)

plt.show() 
```

```python
# You need to compare the data using Line Plot as well as Scatter Plot.

df.plot(x="Name_of_student", y=["Class1 Percentage", "Class2 Percentage","Class3 Percentage"],figsize=(8,8))
a=df["Name_of_student"]
x=df["Class1 Percentage"]
y=df["Class2 Percentage"]
z=df["Class3 Percentage"]

# plt.figure(figsize=(8,8))
plt.scatter(a,x,color="red")
plt.scatter(a,y,color="black")
plt.scatter(a,z,color="blue")

plt.xlabel("Student ",color = "black", weight = "bold", size = 12)
plt.ylabel("percentage",color = "black", weight = "bold", size = 12)
plt.title("Percentage of student",color = "black", weight = "bold", size = 16)

plt.show()
```

```python
# You need to compare the data using there subplots also

import matplotlib.pyplot as plt

x1=df["Name_of_student"]
y1=df["Class1 Percentage"]
plt.subplot(2,2,1)
plt.plot(x1,y1,color="r") 

x2=df["Name_of_student"]
y2=df["Class2 Percentage"]
plt.subplot(2,2,2)
plt.plot(x2,y2,color="g") 

x3=df["Name_of_student"]
y3=df["Class3 Percentage"]
plt.subplot(2,2,3)
plt.bar(x3,y3,color="blue")  #bar
```
